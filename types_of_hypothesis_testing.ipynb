# Two independent sample t-test

```python
import pandas as pd
from scipy import stats
```

```python
data= pd.read_csv('.\Datasets\Mall_Customers.csv')
```

```python
data
```

```python
# seperate data by genere

male_spending=data[data['Genre']=='Male']['Spending Score (1-100)']
female_spending=data[data['Genre']=='Female']['Spending Score (1-100)']

male_spending=male_spending.sample(n=30,replace=False)
female_spending=female_spending.sample(n=30,replace=False)

# perform two sample test
t_ststistic,p_value=stats.ttest_ind(male_spending,female_spending)

# Degree of freedom
n_male = len(male_spending)
n_female = len(female_spending)
degree_of_freedom=n_male-n_female-2
alpha=0.025

critical_value = stats.t.ppf(1-alpha,degree_of_freedom)

print('T-sctatics',t_ststistic)
print('P-value',p_value)
print('Critical Value',critical_value)

if p_value <alpha:
    print('Reject H0')
else:
    print('Fail to reject H0')
```

# Anova


```python
from scipy.stats import f_oneway
from scipy import stats
```

```python
df=pd.read_csv(r'.\Datasets\cust_seg.csv')
```

```python
s1= df.Latest_mon_usage[df.segment == 1]
s2= df.Latest_mon_usage[df.segment == 2]
s3= df.Latest_mon_usage[df.segment == 3]

# Anova
fstate,p_value=f_oneway(s1,s2,s3)
between_df=len([s1,s2,s3]) -1 # number of groups -1
with__df=len(s1) + len(s2) + len(s3) -len([s1,s2,s3]) # total observation - number of groups

alpha= 0.025
critical_value=stats.t.ppf(1-alpha,between_df,with__df)

print(fstate)
print(critical_value)
print(p_value)

if p_value < alpha:
    print("Reject H0")
else:
    print("Fail to reject H0")
```
