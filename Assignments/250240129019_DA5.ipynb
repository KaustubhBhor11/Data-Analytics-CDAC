10. Write a Python program to create bar plot of scores by group and gender.
Use multiple X values on the same chart for men and women.
Sample Data:
Means (men) = (22, 30, 35, 35, 26)
Means (women) = (25, 32, 30, 35, 29)

```python

groups = ['G1', 'G2', 'G3', 'G4', 'G5']
men_means = [22, 30, 35, 35, 26]
women_means = [25, 32, 30, 35, 29]

x = np.arange(len(groups)) 
width = 0.35  

fig, ax = plt.subplots()
bars1 = ax.bar(x - width/2, men_means, width, label='Men', color='green',edgecolor='black')
bars2 = ax.bar(x + width/2, women_means, width, label='Women', color='red',edgecolor='black')

ax.set_xlabel('Groups')
ax.set_ylabel('Scores')
ax.set_title('Scores by group and gender')
ax.set_xticks(x)
ax.set_xticklabels(groups)
ax.legend()
```

13. Write a Python program to create bar plots with errorbars on the same
figure. Attach a text label above each bar displaying men means (integer
value).
Sample Data
Mean velocity: 0.2474, 0.1235, 0.1737, 0.1824
Standard deviation of velocity: 0.3314, 0.2278, 0.2836, 0.2645

```python
velocity = [0.2, 1.2, 2.2, 3.2, 4.2]
scores = [54, 42, 67, 58, 30]
mean_std = [4, 3, 2, 1, 4]
fig, ax = plt.subplots()
bars = ax.bar(velocity, scores, color='red', width=0.5, alpha=0.7, edgecolor='black', yerr=mean_std, capsize=5, label='Men')

for bar, std in zip(bars, mean_std):
    height = bar.get_height()
    ax.text(bar.get_x() + bar.get_width() / 2, height + std + 1, f'{std}', ha='center', va='bottom', fontsize=10, color='k')

plt.xticks(np.arange(0, 5, step=0.5))
plt.title("Scores by Velocity")
plt.xlabel("Velocity")
plt.ylabel("Scores")
plt.legend()

plt.show()
```

14. Write a Python program to create a stacked bar plot with error bars.

15. Write a Python program to create a horizontal bar chart with differently
ordered colors.
Note: Use bottom to stack the women?s bars on top of the men?s bars.
Sample Data Set:
languages = [['Language','Science','Math'],
['Science','Math','Language'],
['Math','Language','Science']]
numbers = [{'Language':75, 'Science':88, 'Math':96},
{'Language':71, 'Science':95, 'Math':92},
{'Language':75, 'Science':90, 'Math':89}]

```python
# Data
students = ['Student 1', 'Student 2', 'Student 3']
numbers = [{'Language': 75, 'Science': 88, 'Math': 96}, 
           {'Science': 95, 'Math': 92, 'Language': 71},  
           {'Math': 89, 'Language': 75, 'Science': 90}]  

language_scores = [numbers[0]['Language'], numbers[1]['Language'], numbers[2]['Language']]
science_scores = [numbers[0]['Science'], numbers[1]['Science'], numbers[2]['Science']]
math_scores = [numbers[0]['Math'], numbers[1]['Math'], numbers[2]['Math']]

y = np.arange(len(students))
fig, ax = plt.subplots(figsize=(4, 3),edgecolor='k')

ax.barh(y[0], language_scores[0], label='Language', color='red',edgecolor='k')
ax.barh(y[0], science_scores[0], left=language_scores[0], color='blue',edgecolor='k')
ax.barh(y[0], math_scores[0], left=language_scores[0] + science_scores[0], color='green',edgecolor='k')

ax.barh(y[1], science_scores[1], color='blue',edgecolor='k')
ax.barh(y[1], math_scores[1], left=science_scores[1], color='green',edgecolor='k')
ax.barh(y[1], language_scores[1], left=science_scores[1] + math_scores[1], color='red',edgecolor='k')


ax.barh(y[2], math_scores[2], color='green',edgecolor='k')
ax.barh(y[2], language_scores[2], left=math_scores[2], color='red',edgecolor='k')
ax.barh(y[2], science_scores[2], left=math_scores[2] + language_scores[2], color='blue',edgecolor='k')


ax.set_yticks(y)
ax.set_yticklabels(students)
ax.set_xlabel('Scores')
ax.legend(['Language', 'Science', 'Math'],bbox_to_anchor=(1, 1))

plt.grid(which='minor',color='g',linestyle=":",alpha=0.3,zorder=1)
plt.grid(which='major',color='g',zorder=1,alpha=0.3)
plt.minorticks_on()
plt.show()
```

16. Write a Python program to create stack bar plot and add label to each
section.


```python
people = ('G1', 'G2', 'G3', 'G4', 'G5', 'G6', 'G7', 'G8')
segments = 4
data = [
    [3.40022085, 7.70632498, 6.4097905, 10.51648577, 7.5330039, 7.1123587, 12.77792868, 3.44773477],
    [11.24811149, 5.03778215, 6.65808464, 12.32220677, 7.45964195, 6.79685302, 7.24578743, 3.69371847],
    [3.94253354, 4.74763549, 11.73529246, 4.6465543, 12.9952182, 4.63832778, 11.16849999, 8.56883433],
    [4.24409799, 12.71746612, 11.3772169, 9.00514257, 10.47084185, 10.97567589, 3.98287652, 8.80552122]
]

data = np.array(data)
totals = data.sum(axis=0) 
y = np.arange(len(people))  

colors = ['red', 'green', 'white', 'purple']


fig, ax = plt.subplots(figsize=(7, 5))
left = np.zeros(len(people))  


for i in range(segments):
    ax.barh(y, data[i], label=f'Segment {i+1}', color=colors[i], left=left,edgecolor='k')
    for j in range(len(people)):
        percent = (data[i][j] / totals[j]) * 100
        ax.text(left[j] + data[i][j] / 2, y[j], f'{percent:.1f}%', ha='center', va='center', fontsize=8, color='black')
    left += data[i]  

ax.set_yticks(y)
ax.set_yticklabels(people)
ax.set_xlabel('Scores')

plt.tight_layout()
plt.show()
```
