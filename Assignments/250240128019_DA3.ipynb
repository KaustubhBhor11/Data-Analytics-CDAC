4. Write a Pandas program to create a new 'Quality -color' Series (use bracket
notation to define the Series name) of the diamonds DataFrame.

```python
df['Color'].value_counts()
def change_color(color):
    dict1={
    'G':'Good',
    'E':'Excellent',
    'F':'Fair',
    'H':'High',
    'D':'Dull',
    'I':'Inferior',
    'J':'Just OK',
    }

    return dict1[color]
df['Quality_color']=df['Color'].map(change_color)
df.head()
```

30. Write a Pandas program to create a side-by-side bar plot of the diamonds
DataFrame.

```python
cut_counts = df['cut'].value_counts()

plot_data = pd.DataFrame({'Cut': cut_counts}).fillna(0)

ax = plot_data.plot(kind='bar', figsize=(12, 7), width=0.8, color=['skyblue', 'orange'])
plt.title('Side-by-Side Bar Plot of Cut and Color', fontsize=16)
plt.xlabel('Categories', fontsize=12)
plt.ylabel('Counts', fontsize=12)
plt.xticks(rotation=45, fontsize=10)
plt.legend(title='Attributes', fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
