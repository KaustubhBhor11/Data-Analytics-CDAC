A life insuranve salesman sell on average 3 life insurance policy per week use poissions law to calculate the probability 

a. in a given week he will sell some policies

b. in a given week he will sell 2 or more policies but not more than five policies

c. Assuming that per week there are 5 working days what is the probability that on a given day he will sell one policy?

d. In the next week he will sell excatly 3 policies.



```python
#A life insuranve salesman sell on average 3 life insurance policy per week use poissions law to calculate the probability 
#a. in a given week he will sell some policies
mu=3
print("the probablit of selling some policies in a week: ",1-poisson.pmf(k=0,mu=mu))

# b. in a given week he will sell 2 or more policies but not more than five policies
print("the probablit of selling some policies in a week: ",sum(poisson.pmf(k=[2,3,4,5],mu=mu)))

# c. Assuming that per week there are 5 working days what is the probability that on a given day he will sell one policy?
print("the probablit of selling some policies in a week: ",(poisson.pmf(k=1,mu=3/5)))

# d. In the next week he will sell excatly 3 policies.
print("the probablit of selling some policies in a week: ",(poisson.pmf(k=3,mu=mu)))
```

```python
# patient average hospital visits duration 5 hoursusing poisson Distribution
import pandas as pd
from scipy.stats import poisson
import  matplotlib.pyplot as plt

df=pd.read_csv(r'.\Datasets\hospital_visits.csv')
df['START'] = pd.to_datetime(df.START)
df['STOP'] = pd.to_datetime(df.STOP)

# calculate the time difference in hours
df['DURATION'] = (df['STOP'] - df['START']).dt.total_seconds() / 3600
print(df)

avg_duration = df['DURATION'].mean()

# define range for poission distribiution (eg around the average)
# will have start and end of the raneas two values
k = range(int(avg_duration) -5 , int(avg_duration)+6) #+/- 5 hours from avg
probaability = poisson.pmf(k,avg_duration)
# print probabilities
print("possion probablities for each k ")
for duration , prob in zip(k,probaability):
    print(f"{duration} hours {prob}")

# plot poission distributoin
plt.figure(figsize=(10,6))
plt.plot(k,probaability,label = 'Poission Distribution',marker='o',color='blue')
plt.xlabel('duration hours')
plt.ylabel('Probability')
plt.title(' Possion distribution for event distribution')
plt.legend()
plt.grid(True)
plt.show()
```

```python

import pandas as pd
from scipy.stats import poisson
import  matplotlib.pyplot as plt

df=pd.read_csv(r'.\Datasets\Retail_Data_Transactions.csv')

df['trans_date'] = pd.to_datetime(df['trans_date'])
transactions_per_day = df.groupby('trans_date').size()
avg_transactions = transactions_per_day.mean()
print(f"Average number of transactions per day: {avg_transactions:.2f}")

# define range for poission distribiution (eg around the average)
# will have start and end of the raneas two values
k = range(int(avg_transactions) -5 , int(avg_transactions)+6) #+/- 5 hours from avg
probaability = poisson.pmf(k,avg_transactions)
# print probabilities
print("possion probablities for each k ")
for duration , prob in zip(k,probaability):
    print(f"{duration} hours {prob}")

# plot poission distributoin
plt.figure(figsize=(10,6))
plt.plot(k,probaability,label = 'Poission Distribution',marker='o',color='blue')
plt.xlabel('Number of transactions')
plt.ylabel('Probability')
plt.title(' Possion distribution for Transaction per Day')
plt.legend()
plt.grid(True)
plt.show()
```
