```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Simulated data : year and coresponding sales
a=np.arange(2010,2021) # years 2010 to 2020
b=np.array([100,120,130,160,180,200,240,260,280,300,320]) # salesdata

# Simulated data: age vs blood pressure
x=np.arange(20,70,5) # ages frp, 20 to 65
y=0.8* x + 80 # blood pressure value
```

```python
# plot 1 Age vs Blood pressure
plt.figure(figsize=(8,5))
plt.plot(x,y,color='green',marker='s',linestyle='-',label='BP vs Age')

```

```python
# plot 2 year vs sales
plt.figure(figsize=(8,5))
plt.plot(a,b,color='blue',marker='s',linestyle='--',label='annaual sales')
plt.title("Annual sales Trends")
```

```python
# pandas vs matplotliv vs seaborn
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('heart.csv')
ages= df.groupby("age").median().reset_index()

# simple pandas
ages.plot.line("age","chol")
plt.show()

# mathplotlib
fig,ax= plt.subplots()
# line style and line width
ax.plot(ages['age'],ages['chol'],linestyle=":",lw=1.7)
ax.set_xlabel("Age")
ax.set_ylabel("Cholesterol")
plt.show()

#seaborn
sns.lineplot(x="age",y='chol',data=ages,linestyle=":",linewidth=1.7)
plt.xlabel("Age")
plt.ylabel("Cholesterol")
```
