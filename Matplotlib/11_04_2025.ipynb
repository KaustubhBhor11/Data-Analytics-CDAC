```python
import numpy as np
import pandas as pd
```

```python
# 
df=pd.read_csv('temp.csv')
```

```python
#Extract Ny column only
my_data=df['NY']

#calculate q1 (25th percentile) and q3 75 th percentile
q1=np.percentile(my_data,25)
q3=np.percentile(my_data,75)
iqr=q3-q1

lower_bound=q1-(1.5*iqr)
upper_bound=q3+(1.5*iqr)

#identify outliers indices
outlier_indices=my_data[(my_data<lower_bound)|(my_data>upper_bound)].index
print(df.loc[outlier_indices])


sns.boxplot(data=my_data,orient='h')
plt.title('temp in NY')
plt.xlabel('temp')
plt.show()
```

```python
df=pd.read_csv('weight-height.csv')
df.head()
```

```python
df_male_height=df.loc[df['Gender'] == 'Male','Height']
df_male_height.head()

q1=np.percentile(df_male_height,25)
q3=np.percentile(df_male_height,75)
iqr=q3-q1
lower_bound=q1-(1.5*iqr)
upper_bound=q3+(1.5*iqr)

outlier_indices=df_male_height[(df_male_height<lower_bound)|(df_male_height>upper_bound)].index
print(len(df_male_height[outlier_indices]))
print(df_male_height[outlier_indices])

sns.boxplot(data=df_male_height,orient='h')
plt.title('Male Hight')
plt.xlabel('Height')
plt.show()
 
```

```python
min(df_male_height),max(df_male_height)
```

```python
# scatter plot sample
plt.figure(figsize=(16,8))
plt.scatter(df_male_height.index,df_male_height,color='blue',alpha=0.5)
plt.title("sctter plot of male height")
plt.xlabel("Index")
plt.ylabel("Height (inches0)")
plt.grid(True)
plt.show()
```

```python
# Histogram to see frequesncy distribution - male height
plt.figure(figsize=(10,6))
plt.hist(df_male_height,bins=20,color='g',edgecolor="black",histtype='barstacked')
plt.title("histogram of male height")
plt.xlabel("Index")
plt.ylabel("Height (inches)")
plt.grid(True)
plt.show()
```
