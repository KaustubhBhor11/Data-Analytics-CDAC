3. Write a Python program to draw a line using given axis values taken from a
text file, with suitable label in the x axis, y axis and a title.
Test Data:
test.txt

1 2

2 4

3 1


```python
with open(r'test.txt') as f :
    pts=f.read()
    f.close()
    x=[]
    y=[]
    for i in pts.split("\n") :
        x_y=i.split(" ")
        print(x_y)
        x.append(int(x_y[0]))
        y.append(int(x_y[1]))
    plt.plot(x,y)
    plt.title('Draw a graph')
    plt.xlabel('x-axis')
    plt.ylabel('y-axis')
    plt.show()

```

11. Write a Python program to plot several lines with different format styles in
one command using arrays.

```python

x = np.linspace(0, 5, 35)
y1 = x**2
y2 = x**3
y3 = x**4

# Plot multiple lines with different styles
plt.scatter(x, y1, color='r', label='y = x^2',s=6)  # Red dashed line
plt.scatter(x, y2, color='g', label='y = x^3',s=12)  # Green dotted line
plt.scatter(x, y3, color='b', label='y = x^4',s=1)  # Blue dash-dot line

# Add title, labels, and legend
plt.title("Multiple Lines with Different Styles")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
# Show the plot
plt.show()
```

14. Write a Python program to display the grid and draw line charts of the
closing value of Alphabet Inc. between October 3, 2016 to October 7, 2016.
Customized the grid lines with rendering with a larger grid (major grid) and a
smaller grid (minor grid).Turn on the grid but turn off ticks.

```python
x1=['2016-10-03','2016-10-04','2016-10-05','2016-10-06','2016-10-07']
y1=[772.5]
df=pd.read_csv('datess.csv')
x=list(df['Date'])
y=list(df['Close'])
plt.plot(x,y,color='red',marker="o",mec='k',mfc='r')
plt.title('Draw a graph')
plt.xlabel('Date')
plt.ylabel('Closing Value')
plt.title('Closing stock value of alphabet Inc.')
plt.grid(color='black',which='both',linestyle=':',alpha=0.5)
plt.grid(which='major',color='red',linestyle="-",alpha=0.7)
plt.minorticks_on()
plt.show()
```

15. Write a Python program to create multiple plots.

```python
import matplotlib.pyplot as plt

# Create a figure
plt.figure(figsize=(10, 8))

# Large plot on top
plt.subplot2grid((3, 3), (0, 0), colspan=3)  # Spans all 3 columns

# First small plot (bottom left)
plt.subplot2grid((3, 3), (1, 0))  # 1st row, 1st column

# Second small plot (bottom center)
plt.subplot2grid((3, 3), (1, 1))  # 1st row, 2nd column

# Third small plot (bottom right)
plt.subplot2grid((3, 3), (1, 2))  # 1st row, 3rd column

# Adjust layout to prevent overlap
plt.tight_layout()
plt.axis
# Show the plots
plt.show()
```
