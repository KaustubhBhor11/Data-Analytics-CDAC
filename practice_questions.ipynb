## Q1. Apply z-test test to check if Monthly Income of male and females are significantly different or same.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy import stats
from statsmodels.stats.weightstats import ztest
df= pd.read_csv(r'.\Datasets\emp.csv')
df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
male_income=df[df['Gender']=='Male'].MonthlyIncome[:400]
female_income=df[df['Gender']=='Female'].MonthlyIncome[:400]
# print(len(male_income),len(female_income))


z_statistic,p_val=ztest(male_income,female_income)
print("Sample size:", len(male_income) + len(female_income))
print("Mean income of males:", np.mean(male_income))
print("Mean income of females:", np.mean(female_income))
print("Standard deviation of male incomes:", np.std(male_income))
print("Standard deviation of female incomes:", np.std(female_income))
print("z-statistic:", z_statistic)
print("p-value:", p_val)

alpha = 0.05
if p_val < alpha:
    print("Reject the null hypothesis: there is a significant difference in monthly income between males and females.")
else:
    print("Fail to reject the null hypothesis: there is no significant difference in monthly income between males and females.")


plt.hist([male_income, female_income], alpha=0.7, label=['Male', 'Female'])
plt.xlabel('Income')
plt.ylabel('Frequency')
plt.title('Distribution of Male and Female Incomes')
plt.legend()
plt.show()

import matplotlib.pyplot as plt

plt.boxplot([male_income, female_income], vert=False)
plt.title('Boxplot of Male and Female Incomes')
plt.show()
```

b. Outliers in AQI of cyprus

```python
cyprus_df=aqi_df[aqi_df['Country']=='Cyprus']
q1=np.quantile(cyprus_df['AQI Value'],0.25)
q3=np.quantile(cyprus_df['AQI Value'],0.75)

iqr=q3-q1

lower_bound=q1-(1.5*iqr)
upper_bound=q3+(1.5*iqr)

outliers=cyprus_df[(cyprus_df['AQI Value']<lower_bound)|(cyprus_df['AQI Value']>upper_bound)]
outliers

print(lower_bound,upper_bound)
```
