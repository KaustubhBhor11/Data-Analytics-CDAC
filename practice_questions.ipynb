d. Perform label encoding and one-hot encoding for relevent columns

```python
from sklearn.preprocessing import LabelEncoder,OneHotEncoder

label_encoder = LabelEncoder()
df['Gender_Encoded'] = label_encoder.fit_transform(df['Gender'])

onehot_encoder=OneHotEncoder(sparse=False)
department_encoded = onehot_encoder.fit_transform(df[['Department']])

if hasattr(onehot_encoder, 'get_feature_names_out'):
    feature_names = onehot_encoder.get_feature_names_out(['Department'])
else:
    feature_names = [f'Color_{cat}' for cat in onehot_encoder.categories_[0]]
df_dept_encoded = pd.DataFrame(department_encoded, columns=feature_names)
```
